Observe que você terá que binarizar a imagem: gere o seu histograma e escolha um valor de limiar para mapear, acima dele, no nível de cinza 255; e abaixo ou igual a ele, no nível de cinza 0.

```python
# Leitura da imagem normal

img = mpimg.imread('Fig8.02.jpg') #lendo imagem
print(img) #print da imagem em matrix
imgplot = plt.imshow(img, cmap ='gray') #print da imagem como figura
```

```python
plt.hist(img) #funcao de histograma
```

```python
#transformando imagem em binaria

imgb = np.zeros((558,558)) #criando vetor vazio do mesmo tamanho da imagem
for i in range(len(img)):
    for j in range(len(img)):
        if img[i][j] <= 127:
            imgb[i][j] = 0
        else:
            imgb[i][j] = 255
        
imgplot = plt.imshow(imgb, cmap = 'gray') #print da imagem binaria
```
